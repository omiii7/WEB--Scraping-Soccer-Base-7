```python
def getLink(id1,id2):
    return 'https://www.soccerbase.com/teams/head_to_head.sd?team_id='+str(id1)+'&team2_id='+str(id2)
#create a funtion for passsing the links for the team id selected.

import requests
from bs4 import BeautifulSoup as bs
import pandas as pd
#import req lib.

url1='https://www.soccerbase.com/teams/home.sd'
response1=requests.get(url1)
soup=bs(response1.text)
#check the connection and cob=nvert the data into text format in soup obj.


import re
s=soup.find_all("a", href = re.compile(r'/teams/team.sd\?team_id=\w+&'))
# for i in s:
#     print(i.text)
menu={i.text :  re.findall('team_id=(\d+)',i['href'])[0] for i in s}
print('SELECT THE IDS OF THE 2 TEAMS : from the menu')
print()
print(menu)
#import re for finding the pattern of the links.

url = getLink(input(),input())
response=requests.get(url)
response
#select the the team id from menu and give the input here.

soup=bs(response.text,'html.parser')
soup

main_lst = []
#create a main list here to append all the list into it .

years=[]
for i in soup.find_all(class_="dateTime"):   
    years.append(i.text[-4:])
main_lst.append(years)

Trophies=[]
for i in soup.find_all(class_="first tournament"):
    Trophies.append(i.text)
main_lst.append(Trophies)

date_t=[]
for i in soup.find_all(class_="dateTime"):
    date_t.append(i.text)
main_lst.append(date_t)

home_team=[]
for i in soup.find_all(class_="team homeTeam"):
      home_team.append(i.text)
main_lst.append(home_team)

away_team=[]
for i in soup.find_all(class_="team awayTeam"):
      away_team.append(i.text)
main_lst.append(away_team)

ht_goals=[]
for  i in soup.find_all(class_="vs"):
    ht_goals.append(i.text[0])
main_lst.append(ht_goals)

at_goals=[]
for  i in soup.find_all(class_="vs"):
    at_goals.append(i.text[-1])
# print(at_goals)
main_lst.append(at_goals)


df = pd.DataFrame(main_lst).T
df.columns=['years','trophies','date','home_team', 'away_team', 'ht_goals', 'at_goals']
#Convert all the row format into columns using the transpose.
# define the col names as per req

df['ht_goals'] = pd.to_numeric(df['ht_goals'], errors='coerce').fillna(0).astype(int)
df['at_goals'] = pd.to_numeric(df['at_goals'], errors='coerce').fillna(0).astype(int)
#convert the data into integer format and fill nan val as zero.

df.loc[((df["ht_goals"]==1) & (df["at_goals"]==1)), "result"]="draw"
df.loc[((df["ht_goals"]==0) & (df["at_goals"]==0)), "result"]="draw"
df.loc[(df['ht_goals']>df['at_goals']),'result']='Home Win'
df.loc[(df['at_goals']>df['ht_goals']),'result']='Away Win'
#Give the conditions requred to convert the data into data frame
print()
print('DATAFRAM OF THE 2 SELECTED TEAMS:- ')
df.head()
#the output :-
```
